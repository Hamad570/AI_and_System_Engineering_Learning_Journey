```python
#9.Create method to update object values.
class employee:
    def __init__(self,name,salary):
        self.name=name
        self.salary=salary
    def updated_values(self,new_name,new_salary):
        self.name=new_name
        self.salary=new_salary
        print(f"Name: {self.name}\nsalary: {self.salary}")


e1=employee('Ahsan', 50000)
e1.updated_values("ali", 70000)
print(e1)
```

```python
#15.Create class for Student and calculate average marks.
class Student:
    def __init__(self,name,*args):
        self.arguments=args
        
        self.sum=0
        for i in args:
            arg=int(i)
            self.sum+=arg
        print(f"The student{name} have total Marks: {self.sum}")
    def grades(self,total):
        self.total=total
        total_n=int(total)
        total_num=(self.sum/total_n)*100
        print(f"Percentage:{total_num}%")
        if total_num>=90:
            print(f"Grade:A+")
        elif total_num>=80:
            print("Grade:A")
        elif total_num>=75:
            print("Grade:B")
        elif total_num>=60:
            print("Grade:C")
        elif total_num>=50:
            print("Grade:D\nPassed")
        else: print("Fail")
s1=Student("ALi",23,24,24,22,21,24)
s1.grades(150)
```

```python
#16.Create class for Temperature converter.
class Temp:
    def  __init__(self,t,scale):
        self.temperature=t
        self.scale=scale
        self.new_tep=0
        if (scale=='F'):
            self.new_tep=(t-32)/1.8
            print(f"Temperature {t}{scale} is {self.new_tep}C in celsius.")
        elif (scale=='C'or'c'):
            self.new_tep=(t*1.8)+32
            print(f"Temperature {t}{scale} is {self.new_tep}F in Forienhite.")
        else: print("Error: Invalid Scale!.")

t1=Temp(98.60000000000001,'F')
t1
```
