# Constructors & Object Methods

```python
#1.Create default constructor example.
class employee:
    def __init__(self):
        self.name='Ali'
        self.salary=25000
    def __str__(self):
       return f"{self.name} have salary {self.salary}."
e1=employee()
e2=e1
print(e1,e2)
```

```python
#2.Create parameterized constructor example.
class employee:
    def __init__(self,name,salary):
        #3.Create instance variable example.(instance varriable = Constructor ke andar)
        self.name=name
        self.salary=salary
    def __str__(self):
       return f"{self.name} have salary {self.salary}."
e1=employee('Ahsan',50000)
print(e1)
```

```python
#4.Create class with multiple methods.
#Methods here = Functions in Class.
class BankAccount:
    def __init__(self,name,am):
        self.am=am
        self.name=name
        self.am=int(am)
    #method 1
    def deposit(self,amo):
        if amo>0:
           self.amo=amo
           print("\n===Deposit Detail====")
           print(f"Previous Balance:{self.am}")
           self.am=self.am+amo
           print(f"The amount {amo} is deposited to the account.")
           print("total balance is: ",self.am)
        else: print("\n===Deposit Detail====\nError: Invalid amount!")
    #method 2
    def withdraw(self,wit):
        if wit<self.am:
            self.wit=wit
            self.am=self.am-wit
            print("\n===Withdraw Detail===")
            print(f"Amount {wit} is withdrawn.\nNew Balance: {self.am}")
        else:
            print("\n===Withdraw Detail===\nError: Insufficient Balance!")
    #5..Create method to compare two objects.
    def comp(self,acc):
        if self.am>acc.am:
            print(f"\n===comparison Detail===\n{self.name} ke paas {acc.name} se zyada paise hain.")
        else: print(f"\n===comparison Detail===\n{acc.name} ke paas {self.name} se zyada paise hain.")

    def display(self):
        print("\n====Account Detail=====")
        print(f"Account Holder: {self.name}")
        print(f"Current Balance: {self.am}")
#6.Print object details using method.
#3.Create instance variable example.
ac1=BankAccount('Ali',4000)
ac2=BankAccount("Ahemd",3000)
ac1.comp(ac2)
ac1.deposit(10000)
ac1.withdraw(600)
ac1.deposit(200)
ac1.display()
```

```python
#7.Create destructor example.
class employee:
    def __init__(self,name,salary):
        #3.Create instance variable example.(instance varriable = Constructor ke andar)
        self.name=name
        self.salary=salary
    def __str__(self):
       return f"{self.name} have salary {self.salary}."
    def __del___(self):
        print(f"--- [Destructor]: Close file '{self.name}' and free memory ---")
e1=employee('Ahsan',50000)
print(e1) 
```

```python
class Student:
    # 8. Class Variable (Shared by all objects)
    student_count = 0 

    def __init__(self, name):
        self.name = name  # Instance variable (Unique for each student)
        
        # 9. Incrementing the class variable using the Class Name
        Student.student_count += 1 

    def display_details(self):
        print(f"Student Name: {self.name}")

print(f"Initial Count: {Student.student_count}")
# Creating 3 different objects
s1 = Student("Ali")
s2 = Student("Zain")
s3 = Student("Ayesha")

# Accessing the shared class variable
print("\n--- Total Objects Created ---")
print(f"Total Students: {Student.student_count}") 

# You can also access the class variable through any object
print(f"Accessed via s1: {s1.student_count}")
print(f"Accessed via s3: {s3.student_count}")
```
