# Inheritance

```python
#Create parent and child class.
#Demonstrate single inheritance.
class parent():
    def __init__(self):
        print("Parent CLass Code!.")
class child(parent):
      print("Child Class Code!")
      pass
obj=child()
```

```python
#Inherit properties from parent class.
class parent():
    def __init__(self,hair,eyes):
        self.hairs=hair
        self.eyes=eyes
    def display(self):
        print("Hairs:",self.hairs)
        print("Eyes: ",self.eyes)
        print("Parent class Code!.")
class child(parent):
    pass

obj=child("Brown","Green")
obj.display()
```

```python
#Create Vehicle class and Car child class.
class veichle():
    def __init__(self,model,brand):
        self.brand=brand
        self.model=model

        print(f"Model: {model}\nBrand: {brand}")

class car(veichle):
    def __init__(self, model, brand,color,engine_power):
        #Use super() to call parent constructor.
        super().__init__(model, brand)
        self.color=color
        self.engine_power=engine_power

        print(f"Color: {color}\nEngine: {engine_power}KW")

c=veichle("Toyota",2024)
print("\n")
b=car(2025,"ferrari","Black",500)
        
```

```python
#Create Animal parent class and Dog child class.
class animal():
    def __init__(self,name):
        self.name=name
        print(f"Name of specie: {name}")
class dog(animal):
    def __init__(self, name,voice,types):
        super().__init__(name)
        self.voice=voice
        self.types=types
        print(f"Voice of {name} is: {voice}\nNumber of Types: {types}")
a=animal("DOG")
print("\n")
b=dog("DOG","Bark",25)
```

```python
#Create Person class and Student child class.
class person():
    def __init__(self,name,number):
        self.name=name
        self.number=number
    def display(self):
        return f"Name:{self.name}\nColor:{self.number}"
class student(person):
    def __init__(self, name,number,RollNO,Class):
        super().__init__(name,number)
        self.rollNo=RollNO
        self.Class=Class
    def display(self):
        print(person.display(self),f"\nRollNO: {self.rollNo}\nClass: {self.Class}")
#Create Teacher class inheriting Person class.
class teacher(person):
    def __init__(self, name, number,Qualification,salary):
        super().__init__(name, number)
        self.salary=salary
        self.Qualification=Qualification
    def display(self):
        print(person.display(self))
        print(f"Qualification: {self.Qualification}\nSalary: {self.salary}")



p=person("ALI",+92345678)
student=student("AHmed",+9233243234445,12,"4th")
student.display()
teacher=teacher("Usama",+92345765783,"M.phil",50000)

teacher.display()
```

**Demonstrate multilevel inheritance**

```python
class animal():
    def __init__(self,name):
        self.name=name
class Dog(animal):
    def __init__(self,name,voice,types):
        super().__init__(name)
        self.voice=voice
        self.types=types
        print(f"\nThe name of Animal is: {name}\nVoice: {voice}\nTypes:{types}")
class German(Dog):
    def __init__(self, name, voice, types,Height,HairColor):
        super().__init__(name, voice, types)
        self.Height=Height
        self.HairColor=HairColor
        print(f"Height:{Height}\nHairColor: {HairColor}")

d=German("Dog","Barking","German","3feet","Brown")
```

**Demonstrate hierarchical inheritance**

```python
# Superclass
class Vehicle:
    def honk(self):
        print("Beep beep! The vehicle is honking.")
# Subclass 1
class Car(Vehicle):
    def open_trunk(self):
        print("Trunk opened.")
# Subclass 2
class Motorcycle(Vehicle):
    def do_wheelie(self):
        print("Doing a wheelie!")
my_car = Car()
my_car.honk()         
my_car.open_trunk()   
my_bike = Motorcycle()
my_bike.honk()        
my_bike.do_wheelie()  
```

**Multiple Inheritance**

```python
class A:
    def __init__(self):
        super().__init__()
        print("Initializing A")

class B:
    def __init__(self):
        super().__init__()
        print("Initializing B")

class C(A, B):
    def __init__(self):
        super().__init__()
        print("Initializing C")

obj = C()
```
