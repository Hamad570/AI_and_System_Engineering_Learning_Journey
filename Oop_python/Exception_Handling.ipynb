```python
#1.division by zero error.
try:
   num1=3
   result1=10/num1
   num2=0
   result2=10/num2
except ZeroDivisionError:
    print("Cannot be divided by 0")
else:
    print(f"result: {result1}")
    #print(f"result2:{result2}")
finally:
    print("Ended Program.")
```

```python
#2.invalid integer input.
try:
   num='ALI'
   num=int(num)
except ValueError:
    print("ValueError: typed wrong")
else:
    print(f"result: {num}")
finally:
    print("Ended Program.")
```

```python
#3.file not found error
try:
   nam={'name':'ALI'}
   print(nam['age'])
except KeyError:
    print("ValueError: Not Found")
else:
    print(f"result: {nam}")
finally:
    print("Ended Program.")
```

```python
#4.index out of range error.
try:
   nam=[1,2,'l']
   print(nam[4])
except IndexError:
    print("ValueError: index out of range")
#Use else block with try-except.
else:
    print(f"result: {nam}")
#Use finally block example.
finally:
    print("Ended Program.")
```

```python
#5.Handle key error in dictionary.
dict={
    "name":"Hamad Hussain",
    "Roll_no":259,
    "MArks":23
}
print("Dictionary=",dict)
#key error occurs when we access that key which doesnt exist.
try:
    print(dict['Semester'])
except KeyError:
    dict['Semester']=4
    print("Semester:" ,dict['Semester'])
    print("New dictionary=",dict)
```

```python
#6.Handle typeerror in operations.
#An operation or function is applied to an object of an inappropriate or mismatched data type
#Like Result="Age"+12

try:
    name="Hamad"
    roll_no=12
    info=name+roll_no
    print(info)
except TypeError:
    print("TypeError")
```

```python
#Use multiple except blocks.
try:
    #num=int(input("Enter a number: ")) for example:10
    #num2=int(input("Enter 2nd number: ")) like 0
    num=10
    num2=0
    div=num/num2
except ZeroDivisionError:
    print("Numbers cannit be divisible by 0")
except TypeError:
    print("Invalid 2nd number.")
```
