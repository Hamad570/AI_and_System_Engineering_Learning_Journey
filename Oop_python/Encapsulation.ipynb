# Encapsulation

```python
#Create private variable in class.
#Create class with protected variable.
class student:
    def __init__(self,name,age,marks):
        self.name=name #public varriable
        self._age=age   #protected Varriable
        self.__marks=marks #private Varriable

s1=student('Hamad',20,25)
print(s1.name,"Public you can edit this attribute.")
print(s1._age,"Protected parent and child class access")
print(s1.marks,"Private")
```

```python
#Access private variable using method.
class student:
    def __init__(self,name,age):
        self.__name=name
        self.__age=age

    def display(self):
        print(f"student Name: {self.__name}\nAge: {self.__age}")
s1=student('Hamad',20)
s1.display()  
```

```python
#Hide salary details using encapsulation.
class employee:
    def __init__(self,name,salary):
        self.name=name
        self.__salary=salary
    def display(self):
        print("Name: ",self.name)
e1=employee("ALI",5000)
print(e1.display())
        
```

```python

#Create getter and setter methods.
class student:
    def __init__(self,name,age):
        self.__name=name
        self.__age=age
        #GETTER
    @property
    def display(self):
        return self.__name,self.__age
    #SETTER
    @display.setter
    def display(self,n_name):
        self.__name=n_name
        print("Name:",n_name)
s1=student('Hamad',20)
s1.display="Hamad Hussain"
s1.display
```

```python
#Create secure bank account class.
class Bank:
    def __init__(self,Acount,Balance,pin):
        self.Account=Acount
        #Create class with private balance attribute.
        self.__balance=Balance
        self.__pin=pin
    def display(self,pin):
        print(f"The Account {self.Account}")
        if self.__pin==pin:
            print(f"The balance is {self.__balance}")
A1=Bank(12345,5000,7654)
A1.display(764)        
```

```python
#Prevent direct modification of data.
class employee:
    def __init__(self,id,name,salary,rank):
        self.id=id
        self.name=name
        self.__salary=salary
        self.__rank=rank
    @property
    def display(self,):
        print(f"ID:{self.id}\nNAME: {self.name}")
        return self.__rank,self.__salary
    @display.setter
    def display(self,data):
        n_id,n_name=data
        print(f"DATA of Employee {self.id},{self.name} is UPDATED.")
        self.id=n_id
        self.name=n_name
        print(f"===DATA===\nID: {self.id}\nNAME: {self.name}")

e1=employee(1,"Ahmed",50000,"2nd")
e1.display=(101,"M.Ahmed") 
e1.display    
```

```python
#Create login system using private password.
class register:
    def __init__(self,uname,passw):
        self.__password=passw
        self.__name=uname

    def login(self,n_name,password):
        if self.__name==n_name and self.__password==password:
            print("Welcome to the WEbsite....")
        else:
            print("NOT VALID")

a=register('HAMAD',123456)
a.login("HAMAD",123456)
a.login("Hamad",123456)


        
```
