# Abstraction

```python
from abc import ABC,abstractmethod
```

```python
#Create abstract class example.
#Create abstract method using abc module.
class veichle():
    @abstractmethod
    def engine(self):
        pass
    def horn(self):
        print("beep! \n beep!")
class car(veichle):

    def engine(self):
        print("Car Engine is heavy")

class bike(veichle):
    def engine(self):
        print("Bike engine is lighter than cars")


car1=car()
car1.engine()
car1.horn()

bike1=bike()
bike1.engine()
bike1.horn()
```

```python
#Create Shape abstract class with area method.
class Shape(ABC):
    @abstractmethod
    def area(self):
        """Calculate and return the area of the shape."""
        pass
```

```python
#Create Payment abstract class.
class Payment(ABC):
    @abstractmethod
    def process_payment(self, amount: float):
        """Process a transaction for the given amount."""
        pass
```

```python
#Create abstract Animal class.
class Animal(ABC):
    @abstractmethod
    def make_sound(self):
        """Emit the sound characteristic of this animal."""
        pass

    @abstractmethod
    def move(self):
        """Define how the animal moves around."""
        pass
```

```python
#Create abstract database connection system.
class DatabaseConnection(ABC):
    @abstractmethod
    def connect(self, connection_string: str):
        """Establish a connection to the database."""
        pass

    @abstractmethod
    def execute_query(self, query: str):
        """Execute an SQL or NoSQL query against the database."""
        pass

    @abstractmethod
    def disconnect(self):
        """Safely close the database connection."""
        pass
class MySQLConnection(DatabaseConnection):
    def __init__(self):
        self.is_connected = False

    def connect(self, connection_string: str):
        self.is_connected = True
        return f"Successfully connected to MySQL database using: {connection_string}"

    def execute_query(self, query: str):
        if self.is_connected:
            return f"Executing query '{query}' in MySQL and fetching results..."
        return "Error: Cannot execute query. Database is not connected."

    def disconnect(self):
        self.is_connected = False
        return "MySQL Database disconnected safely."

# --- PROGRAM RUN AND OUTPUT ---
db = MySQLConnection()

print(db.connect("mysql://localhost:3306/my_app_db"))
print(db.execute_query("SELECT * FROM users;"))
print(db.disconnect())
```
