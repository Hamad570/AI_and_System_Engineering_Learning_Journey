# Tuples

```python
#1. Create a tuple and print its elements.
t1=(1,2,3,4,1,1,5)
print(t1)
```

```python
#Count occurrences of an item in tuple.
t1.count(1)
```

```python
#Find index of an element in tuple.
t1.index(5)
```

```python
#Convert tuple into list.
l1=list(t1)
print(l1)
# covert list into tuple
t2=tuple(l1)
print(t2)
```

```python
#Find maximum and minimum value in tuple.
print(max(t1))
print(min(t1))
```

# Dictionaries

```python
#Create a dictionary of student information.
students={
    1:{"name":"Hamad","rollno":259},
    2:{"name":"usman","rollno":259}
}

print(students)
print(students[1])
for values in students.values():
 print(values["name"])

print(students.get(1))
```

```python
#Print all keys of dictionary.
for keys in students.keys():
    print(keys)
#Print all values of dictionary.
for values in students.values():
 print(values)
```

```python
#Add a new key-value pair in dictionary.
students.update({3:{"name":"Ali","rollno":280}})
students
```

```python
#Update value of existing key.
students[2]["rollno"]=260
students
```

```python
#Remove a key from dictionary.
#students.popitem()
students.pop(1)
students[2].pop("name")
students
```

```python
#Check if a key exists in dictionary.
def fi(ke):
    if ke in students:
        print(f"student {ke} is found \n --details-- \n {students[ke]}")
    else: print(f"students {ke} NOT FOUND!")
s=3
fi(s)
```
