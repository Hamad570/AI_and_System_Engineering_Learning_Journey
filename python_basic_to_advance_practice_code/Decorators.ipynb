```python
#Basic Calculators Using *args
def add(*args):
    result=0
    for i in args:
        result+=i
    return result
def multiply(*args):
    result2=1
    for i in args:
        result2*=i
    return result2
def sub(*args):
    re=1
    for i in args:
        re=i-re
    return re

add(1,2,3,4,5)
print(multiply(1,2,3,1,1,1,13,4,5,1,1,1))
#sub(10,3,4,2,4)
```

```python
#Greeting Logger (Nested Function)
def greet(name):
    def msg():
        return f"Assalam alaikum {name}, how are you?"
    welcome=msg()
    print(welcome)

greet("Hamad")
```

```python
#Returning Functions from Functions
def power(factor):
    def num(num):
        return num**factor
    return num
double=power(3)
print(double(5))
num1=power(8)
num1(2)
```

```python
def user(name,**kwargs):
    print(f"--profile of {name}--")
    for key,value in kwargs.items():
        print(f"{key.upper()}:{value}")
user("Hamad",age=21,city="Sakhi sarwar",Country="Pakistan",Degree="BSCS",semester="4th")
```

```python
#Passing Function as Argument
def add(*args):
    result=0
    for i in args:
        result+=i
    return result
def multiply(*args):
    result=1
    for i in args:
        result=i*result
        return result
def calculate(func,*args):
    return func(*args)

calculate(add,1,2,3,6,4,5,5)
calculate(multiply,10,2,3,4,5)
```

```python
#Custom Order Invoice (Mix of standard, *args, kwargs)
def order(name,*items,**discount):
    print(f"--{name}--")
    print(f"items: {items}")
    count=0
    for i in items:
        count+=1
    print(f"Total Items:{count}")
    print(f"Discount:{discount}")
    def total(price,discount):
        discounted_price=(discount/100)*price
        price2= price-discounted_price
        print(f"discounted price:{price2}")
    return total
    
order1=order("Hamad","iphone","Mouse","printer","keyboard",discount=10)
print(order1(900000,10))
```

```python
def student(name,*marks,**detail):
     print(f"--{name} Report--")
     print(f"Name: {name}")
     for key,value in detail.items():
        print(f"{key}: {value}")
     def calculator():
        total_marks=0
        for i in marks:
            total_marks+=i
            print(f"Subject Marks:{i}")
        print(f"Total marks:{total_marks}")
     return calculator
s1=student("Hamad",20,30,21,25,34,degree="BSCS",semester="4th",Section="A",)
print(s1())
```

```python
#Nested + Returning Function
def factory(case): 
    def uppercase(text): return text.upper()
    def lowercase(text): return text.lower()

    if case=="upp":
        return uppercase
    else: return lowercase 

fo=factory("upp")
fo("hello")
```

# Decorators

```python
def greet(func):
    def inst(*args):
        print("Hello!")
        ot=func(*args)
        print("Thanks")
        return ot
    return inst
@greet
def marks(*args):
    result=0
    for i in args:
        result+=i
    print(result)
marks(2.3,5)
marks(234,5,0,8,9,7,6)
```

```python
def uppercase(func):
    def wrp():
        result=func()
        return result.upper()
    return wrp
@uppercase
def greet():
    return f"hello kesy ho?"
print(greet())
```
