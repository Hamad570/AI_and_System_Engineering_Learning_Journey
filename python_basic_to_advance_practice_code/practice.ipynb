**Functions**

```python
#1.Create a function to add two numbers.
def sum(a,b):
    return a+b
print(sum(2,4))
#2.Create a function to find square of a number.
def square(a):
    return a**2
print(square(5))
#3.Create a function to check even or odd.
def check(n):
    if n%2==0:
        return "Even"
    else: return "Odd"
print(check(87))
#4.Create a function that returns maximum of three numbers.
def maximum(z,x,y):
    if x>y and x>z:
        return f"{x} is the maximum"
    elif y>x and y>z:
        return f"{y} is the maximum"
    elif z>x and z>y:
        return f"{z} is the largest"
    elif x==y or y==z or x==z:
        return f"The two number are same."
    else: return "Error"
print(maximum(2,3,5))
print(maximum(2,2,0))

#5.Create a function to calculate factorial.
def factorial(n):
    count=1
    for i in range(1,n+1):
        count*=i
    return f"{count} is the factorial of{n}"
print(factorial(10))
#6.Create a function to count vowels in a string.
def vowel(n):
    vowels="aeiouAEIOU"
    result=0
    for i in n:
        if i in vowels:
            result+=1
    return result
print(vowel("hamade"))

#7.Create a function to check palindrome string.
def palin(n):
    n2=n[::-1]
    if n==n2:
        return "Palindrome"
    else: return "Not Palindrome"

print(palin("madamh"))
print(palin("madam"))

#8.Create a function for multiplication table.
def mul(n):
    for i in range(1,11):
      print(f"{n}x{i}={n*i}")
print(mul(5))
#9. Create a function to calculate average of numbers.
def avh(n,m,h):
    return (n+m+h)/3
print(avh(3,2,1))
```
