# File Handling

```python
#Create a text file and write data into it.
file=open('Example.txt','w')
file.write('Hello World \nKesy ho ap?')
file.close()
```

```python
#Read data from a text file.
file=open('Example.txt','r')
print(file.read())#output: Entire data
print(file.readline())# output: show just 1st line
print(file.readlines())#output: pront data in list format
file.close()
```

```python
#Append new data into existing file.
file=open('Example.txt','a')
file.write('\nMaen thk ho.')
file.close()
```

```python
#Count total characters in file.
file=open('Example.txt','r')
characters=len(file.read())
print(characters)
```

```python
#Count total words in file.
file=open('Example.txt','r')
words=file.read()
w=words.split()
count=0
for i in w:
    count+=1
print(f"Total words in file are : {count}")
```

```python
#Count total lines in file.
file=open('Example.txt','r')
w=file.readlines()
count=0
for i in w:
    count+=1
print(f"Total lines in file are : {count}")
```

```python
#Read file line by line.
file=open('Example.txt','r')
w=file.readlines()
count=0
for i in w:
    count+=1
    print(count,":",i)
```

```python
#Copy content from one file to another.
file=open('Example.txt','r')
w1=file.read()
fil2=open('Example2.txt','w')
w2=fil2.write(w1)
file.close()
fil2.close()
```

```python
#Convert file text into uppercase.
file=open('Example.txt','r+')
w=file.read()
file.seek(0)
file.write(w.upper())
file.truncate()
file.close()
```

```python
#Search a word inside file.
file=open('Example.txt','r')
search='ho'
w=file.read()
if search in w:
    print("Found")
else: print("Not Found")
file.close()
```

```python
#Search a word inside file with whole line.
file=open('Example.txt','r')
line_num=1
found=False
line=file.readline()
while line:
    if search.lower() in line.lower():
        print(f"line {line_num}: {line}")
        found=True
    line_num+=1
    line=file.readline()
file.close()
if found==False:
    print('Not FOund')
```

```python
#Replace a word in file.
search='ho'
repl='haen'

file=open('Example.txt','r')
rad=file.read()
r=rad.lower()
file.close()
s=search.lower()
replacement=r.replace(s,repl)
file=open('Example.txt','w')
file.write(replacement)
file.close()
print(f'{search} ko ab {repl} sy replace kr dya gya hai.')
```
