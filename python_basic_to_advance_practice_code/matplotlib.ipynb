```python
import matplotlib.pyplot as plt
import numpy as np
```

# line chart

```python
#Plot a simple line graph.
a=[1,2,3,4,5]
b1=[10,12,14,16,18]
plt.plot(a,b1)
plt.show()
```

# Customization of line chart


```python
#Plot graph with custom figure size.
plt.figure(figsize=(4,4))
plt.legend()
lables=["A","B","c","D","E"]
plt.xticks()
#Change marker style in graph.
# Change line style in graph.
plt.plot(a,b1, color="g",marker="o", linestyle="--",linewidth=2,markersize=12)
#Add title to graph.
plt.title("Simple line graph",color='r')
#Add labels on x-axis and y-axis.
plt.xlabel("a data")
plt.ylabel("b data")
plt.show()
```

```python
# Plot multiple lines on same graph.
b2=[12,15,18,21,24]
plt.Figure(figsize=(5,4))
plt.plot(a,b1,linestyle='--',marker='o',color='red')
plt.plot(a,b2,linestyle='-',marker='x',color='g')
# Plot graph with grid enabled.
plt.grid(True)
plt.title('Year 1 sales')
plt.xlabel('Months')
plt.ylabel('Sales')
plt.show()
```

```python
#Plot graph using random data.
x=[1,2,3,4,5,6,7]
y=[12,15,13,16,16,17,15]
plt.figure(figsize=(6,4))
plt.plot(x,y)
plt.show()
```

# Bar Charts

```python

#Create simple bar chart.
x=['Random','SVM','Linear','Logistic']
y=[0.29,0.99,0.46,0.76]
plt.Figure(figsize=(4,4))
plt.bar(x,y,width=0.5,color='g')
plt.title('Models with Accuracy')
plt.xlabel('Model')
plt.ylabel('Accuracy')
```

```python
#Create horizontal bar chart.
plt.barh(x,y,color='g')
plt.title('Models with Accuracy')
plt.xlabel('Model')
plt.ylabel('Accuracy')
```

```python
#Plot sales data using bar chart.
sales=[20,29,15,9,30]
years=[2022,2023,2024,2025,2026]
plt.figure(figsize=(3,3))
plt.bar(years,sales,color='grey',width=0.25)
plt.title('Sales Report')
plt.xlabel('years')
plt.ylabel('Sales Rate')
plt.show()
```

```python
#Create grouped bar chart.
n=['A','B','C','D']
m1=[200,250,600,100]
m2=[900,400,100,400]
y=np.arange(len(n))
width=0.25
plt.bar(y-(width/2),m1,width,color='Green',label='Sales')
plt.bar(y+(width/2),m2,width,color='Blue',label='buy')
plt.show()
```

```python
#Create stacked bar chart.
plt.barh(n,m1,label='sales',color='red')
plt.barh(n,m2,left=m1,label ='buys',color='grey')
```

```python
#Change width of bars.
plt.bar(n,m1,width=0.25)
#Add labels on bars.
plt.ylabel("sales")
plt.xlabel('Essentials')
plt.show()
```

```python
#Compare student marks using bar chart.
students=['Ali','Usman','MAjeed','REhan']
marks=[23,21,20,25]
#Create colorful category comparison chart.
colors=['red','green','blue','yellow']
plt.bar(students,marks,width=0.35,color=colors)
plt.show()
```

# Pie Charts


```python
#Create simple pie chart.
c=['a','b','c','d']
d=[2,5,1,3]
plt.pie(d,labels=c)
plt.title('Pie chart')
plt.show()
```

```python
#Add percentage labels in pie chart.
colors=['darkblue','Blue','lightBlue','Skyblue']
plt.pie(d,labels=c,colors=colors,autopct='%1.1f%%')
plt.show()
```

```python
#Create pie chart for market share.

shares=[30,20,5,35,3,7]
names=['Ali','Ahmed','Amjad','Amna','Amir','Aima']
plt.pie(shares,labels=names,autopct='%1.1f%%')
plt.title('---The Market Share Holders Share chart-----')
plt.show()
```

```python
#Create pie chart for student grades.
grades=[12,23,17,10,20,25]
plt.pie(grades,labels=names,colors=colors,autopct=lambda x:f'{int(round(x/100.*sum(grades)))}')
plt.title('less than 13 are fail')
plt.show()
```

# Histograms

```python
#Create histogram of random numbers.
y=np.random.randn(1000)
plt.hist(y,edgecolor='black')
plt.title('Historgram of random numbers')
plt.xlabel('Frequency')
plt.ylabel('value')
plt.show()
```

```python
#Plot age distribution using histogram.
a=np.random.randint(15,25,100)
plt.hist(a,bins=10,edgecolor='r')
plt.show()
```

```python
#Change histogram bin size.
b=np.random.randint(1,100,500)
plt.hist(b,bins=50,edgecolor='black')
plt.show()
```

```python
#Compare two histograms.
a=np.random.randint(1,100,400)
plt.hist(a,bins=15,alpha=0.5,edgecolor='red',label='salary',density=True)
plt.hist(b,bins=30,alpha=0.5,edgecolor='black',label='ages',density=True)
plt.show()
```

```python
#Create histogram for exam scores.
scores=np.random.randint(1,25,100)
plt.hist(scores,bins=20,edgecolor='darkgrey',histtype='bar')
plt.show()
```
