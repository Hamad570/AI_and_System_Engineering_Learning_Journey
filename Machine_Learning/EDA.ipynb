```python
import pandas as pd 
import numpy as np
```

```python
#Load dataset using pandas.
data=pd.read_csv('dataset.csv')
```

```python
#Find missing values.
data.isnull()
```

```python
#Count total missing values.
data.isnull().sum()
```

```python
#Check duplicate rows.
print(data.duplicated())
print("Total Duplicated Values: ",data.duplicated().sum())
```

```python
#Remove duplicate rows.
data.drop_duplicates()
```

```python
#Fill missing values with mean.
data=data.fillna(data.mean(numeric_only=True))
```

```python
#Fill missing values with median.
data=data.fillna(data.median(numeric_only=True))
```

```python
#Fill missing values with mode.
data=data.fillna(data.mode(numeric_only=True))
```

```python
#Drop rows with missing values.
data.dropna(how='all')
```

```python
#Rename dataset columns.
data=data.rename(columns={'Student_ID':'ID'})
data.head()
```

```python
#Convert datatype of columns.
data['Semester'].astype('str')
data['Semester'].dtype
```

```python
#Remove unwanted columns.
df=data.drop(columns='ID')
```

```python
#Detect outliers using IQR method.
Q1=df['Age'].quantile(0.25)
Q3=df['Age'].quantile(0.75)
IQR=Q3-Q1
lower=Q1-1.5*IQR
upper=Q3+1.5*IQR
print(f"Lower: {lower}\nUpper: {upper}")
```

```python
#Remove outliers from dataset.
d=df[(df['Age']>=lower)&(df['Age']<=upper)]
```

```python
#Find average marks from dataset.
df['Matric_Marks'].mean()
```

```python
#Find highest and lowest values.
lowest=df['Matric_Marks'].min()
highest=df['Matric_Marks'].max()
print(f"Lowest Marks: {lowest}\nHighest: {highest}")
```

```python
#Count unique values in column.
print(df['Gender'].unique())
```

```python
#Find correlation between columns.
print(df['Matric_Marks'].corr(df['Intermediate_Marks']))
```

```python
#Group data using groupby().
print(df.groupby('Communication_Skill')['Current_CGPA'].mean())
```

```python
#Analyze department-wise Communication_Skill.
print(df.groupby('Department')['Communication_Skill'].mean())
```

```python
#Analyze gender distribution.
df['Gender'].value_counts()
```

```python
#Find top 10 highest values.
#Sort dataset based on column.
data2=df.sort_values('Current_CGPA',ascending=True)
data2.iloc[-11:-1,:]
```
