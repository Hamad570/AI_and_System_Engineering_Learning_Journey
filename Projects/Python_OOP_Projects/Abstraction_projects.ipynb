```python
#Create abstract notification system.
from abc import ABC, abstractmethod
class Notification(ABC):
    def __init__(self, sender, recipient):
        self.sender = sender
        self.recipient = recipient
    @abstractmethod
    def send(self, message):
        pass
class EmailNotification(Notification):
    def __init__(self, sender, recipient, subject):
        Notification.__init__(self, sender, recipient)
        self.subject = subject
    def send(self, message):
        print(f" Sending Email...")
        print(f"From: {self.sender} | To: {self.recipient}")
        print(f"Subject: {self.subject}")
        print(f"Message: {message}\n")
class SMSNotification(Notification):
    def __init__(self, sender, recipient, phone_carrier):
        Notification.__init__(self, sender, recipient)
        self.phone_carrier = phone_carrier
    def send(self, message):
        print(f" Sending SMS Text...")
        print(f"From: {self.sender} | To: {self.recipient}")
        print(f"Carrier: {self.phone_carrier}")
        print(f"Text Body: {message}\n")
email_alert = EmailNotification("bank@secure.com", "user@email.com", "Security Alert")
sms_alert = SMSNotification("+123456", "+987654", "T-Mobile")
email_alert.send("Someone logged into your account.")
sms_alert.send("Your verification code is 4432.")
```

```python
#Create abstract database connection system.
from abc import ABC, abstractmethod
class DatabaseConnection(ABC):
    def __init__(self, host, port, user, password):
        self.host = host
        self.port = port
        self.user = user
        self.password = password
    @abstractmethod
    def connect(self):
        pass
    @abstractmethod
    def disconnect(self):
        pass
class MySQLConnection(DatabaseConnection):
    def __init__(self, host, port, user, password, database_name):
        DatabaseConnection.__init__(self, host, port, user, password)
        self.database_name = database_name
    def connect(self):
        print(f" Connected to MySQL database: {self.database_name} at {self.host}:{self.port}")
    def disconnect(self):
        print(f"Disconnected from MySQL database: {self.database_name}")
class PostgreSQLConnection(DatabaseConnection):
    def __init__(self, host, port, user, password, schema):
        DatabaseConnection.__init__(self, host, port, user, password)
        self.schema = schema
    def connect(self):
        print(f"Connected to PostgreSQL schema: {self.schema} at {self.host}:{self.port}")
    def disconnect(self):
        print(f" Disconnected from PostgreSQL schema: {self.schema}")
mysql_db = MySQLConnection("127.0.0.1", 3306, "root", "pass123", "company_db")
postgres_db = PostgreSQLConnection("192.168.1.1", 5432, "admin", "secure789", "public")
mysql_db.connect()
mysql_db.disconnect()
print("--------------------")
postgres_db.connect()
postgres_db.disconnect()
```
