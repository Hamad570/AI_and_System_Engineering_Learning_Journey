**Multiple Inheritace**


```python
#Create Employee class and Manager child class.
class Employee():
    def __init__(self,name,id,salary):
        self.name=name
        self.id=id
        self.salary=salary
    def display(self):
        print(f"ID: {self.id}\nName: {self.name}\nSalary:{self.salary}")
class Manager(Employee):
    def __init__(self, name, id, salary,daily_tasks):
        Employee.__init__(self,name, id, salary)
        self.daily_tasks=daily_tasks
    def display(self):
        print(Employee.display(self),f"Daily Tasks Completed: {self.daily_tasks}")

class HR(Employee):
    def __init__(self, name, id, salary,Hiring,Invest_management):
        Employee.__init__(self,name, id, salary)   
        self.Hiring=Hiring
        self.Invest_Managent=Invest_management
    def display(self):
        print(Employee.display(self),f"Hirings Open Krein:{self.Hiring}\nInvest Managemnt is ok:{self.Invest_Managent}")

class CEO(HR,Manager):
    def __init__(self, name, id, salary, Hiring, Invest_management,daily_tasks):
        HR.__init__(self,name, id, salary, Hiring, Invest_management)
        Manager.__init__(self,name, id, salary,daily_tasks)
        print(f"ID: {id}\nName: {name}\nSalary:{salary}\nHirings Status:{Hiring}\nInvest_Report:{Invest_management}\nDaily Tasks Completed:{daily_tasks}")

class Founder(CEO):
    def __init__(self, name, id, salary, Hiring, Invest_management, daily_tasks,invest):
        super().__init__(name, id, salary, Hiring, Invest_management, daily_tasks)
        self.invest=invest
        print(f"Investment Amount:{invest}")


ceo=CEO("Ahmed",101,1000000,"NO","YESNeeded","Yes Completed")
print("--------------------\n----------------------")
employee=Employee("ALI",1010,50000).display()
print("------------\n-----------------------------")
manager=Manager("Usman",300,150000,"Completed")
manager.display()
print('---------------------\n-------------------------------')
founder=Founder("Ahmed",101,1000000,"NO","YESNeeded","Yes Completed","7000K more")
```
